```python
# Домашнє завдання. Частина друга
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import seaborn as sns
```

```python
table = pd.read_csv(
    'https://drive.google.com/u/0/uc?id=1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U&export=download')
```

```python
# Визначте розмір таблиці за допомогою методу shape
print(table.shape)
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(table.dtypes)
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_values_ratio = table.isnull().sum() / len(table) * 100
print(missing_values_ratio)
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
table = pd.concat([table.dropna(axis=1), table["Язык.программирования"]], axis=1)
print(table.head())
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing_values_ratio = table.isnull().sum() / len(table) * 100
print(missing_values_ratio)
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
table = table.dropna(axis=0)
print(table.head())
```

```python
#Визначте новий розмір таблиці за допомогою методу shape
print(table.shape)
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = table[table['Язык.программирования'] == 'Python']
print(python_data.head())
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
print(python_data.shape)
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped_data = python_data.groupby('Должность')
for position, group in grouped_data:
    print(f"Должность: {position}")
    print(group.head())  # Вивід перших рядків для кожної групи
    print("------------------")
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
aggregated_data = python_data.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
print(aggregated_data)
```

```python
aggregated_data = python_data.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
labels = np.array(aggregated_data.index)

data = np.array(aggregated_data['Зарплата.в.месяц']['min'])
explode = [0.15 for _ in aggregated_data['Зарплата.в.месяц']['min']]
plt.pie(
    data,
    labels=labels,
    shadow=False,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=0.8,
    labeldistance=1.05,
)

plt.title('Відносна межа найменших зарплат (Python) за посадами.', fontsize=12.5)
plt.show()
```

```python
aggregated_data = python_data.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
labels = np.array(aggregated_data.index)

data = np.array(aggregated_data['Зарплата.в.месяц']['max'])
plt.pie(
    data,
    labels=labels,
    shadow=False,
    autopct="%.2f%%",
    pctdistance=0.6,
    labeldistance=1.1,
)

plt.title('Відносна межа найбільших зарплат (Python) за посадами.', fontsize=12.5)
plt.show()
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(salary):
    return str(int(salary.mean()))

aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
aggregated_data = aggregated_data.reset_index()
aggregated_data.columns = ['Position', 'min', 'max', 'avg']
print(aggregated_data.head())
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
avg_stats = aggregated_data['avg'].describe()
print(avg_stats)
```

```python
aggregated_data.to_csv('salary.csv', index=False)
print(pd.read_csv('salary.csv'))
```
